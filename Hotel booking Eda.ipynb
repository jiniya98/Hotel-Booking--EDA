- INTRODUCTION:
    
    
   The hotel booking dataset contains information on bookings made for two types of hotels: a city hotel and a resort hotel. The dataset includes various features related to the length of stay, the number of available parking spaces,market segments and so on.

   In addition to these features, the dataset also includes information about the hotel itself, such as the most demanded rooms,demanded meals and so on. There is also information about the customers who made the bookings, such as their country of origin and whether they had previously booked with the hotel.

   The dataset is useful for analyzing trends in hotel bookings and customer behavior. For example, the dataset can be used to analyze seasonal trends in hotel bookings or to identify which countries are the largest sources of customers for the hotels. The dataset can also be used to analyze the factors that influence customer decision-making, such as the availability of parking spaces or the length of stay.

   Overall, the hotel booking dataset provides a valuable resource for businesses in the hospitality industry, researchers, and data analysts who want to gain insights into hotel bookings and customer behavior patterns.

###### Data Summary

-Hotel
- 2 hotels are here

-is_canceled
- Value indicating if the booking was canceled (1) or not (0)

-lead_time
- Number of days that elapsed between the entering date of the booking and the arrival date

-arrival_date_year
- arrival year of the guest

-arrival_date_month 
- Month of arrival date with 12 categories: “January” to “December”

-arrival_date_week_number 
- Week number of the arrival date

-arrival_date_day_of_month 
- Day of the month of the arrival date

-stays_in_weekend_nights 
- Number of weekend nights (Saturday or Sunday) the guest stayed or booked to stay at the hotel

-stays_in_week_nights
- Number of week nights (Monday to Friday) the guest stayed or booked to stay at the hotel 

-adults
- Number of adults

-children
- Number of children

-babies
- Number of babies

-Meal
- Type of meal booked

-country
- Country of orgin

-market_segment
- Market segment designation. In categories, the term “TA” means “Travel Agents” and “TO” means “Tour Operators”

-distribution_channel
- The medium through booking was made[Direct,coperate,TA/TO, Undefined,GDS]

-is_repeated_guest
- Shows if the repeated guest is who has arrived or not

-previous_cancellations
- Number of previous bookings that were cancelled by the customer prior to the current booking

-previous_bookings_not_canceled
- Number of previous bookings not cancelled by the customer prior to the current booking

-reserved_room_type
- Code of room type reserved

-assigned_room_type
- Code for the type of room assigned to the booking. Sometimes the assigned room type differs from the reserved room type.

-booking_changes
- Number of changes/amendments made to the booking from the moment the booking was entered on the PMS until the moment of check-in or cancellation

-deposit_type
- No Deposit – no deposit was made; Non Refund – a deposit was made in the value of the total stay cost; Refundable – a deposit was made with a value under the total cost of stay.

-agent
- ID of the travel agency that made the booking

-company
- ID of the company/entity that made the booking or responsible for paying the booking. ID is presented instead of designation for anonymity reasons

-days_in_waiting_list
- Number of days the booking was in the waiting list before it was confirmed to the customer

-customer_type
- Group – when the booking is associated to a group; Transient – when the booking is not part of a group or contract, and is not associated to other transient booking; Transient-party – when the booking is transient, but is associated to at least other transient booking

-adr
- Average Daily Rate (Calculated by dividing the sum of all lodging transactions by the total number of staying nights)

-required_car_parking_spaces
- Number of car parking spaces required by the customer

-total_of_special_requests
- Number of special requests made by the customer (e.g. twin bed or high floor)

-reservation_status
- Check-Out – customer has checked in but already departed; No-Show – customer did not check-in and did inform the hotel of the reason why

-reservation_status_date
- Date at which the last status was set. This variable can be used in conjunction with the ReservationStatus to understand when was the booking canceled or when did the customer checked-out of the hotel

### Combining columns and creating a new column

```python
df['children'].fillna(method='ffill',inplace=True)
```

```python
df['Total Guest']=df['adults']+df['children']+df['babies']
df['total stay in nights']=df['stays_in_weekend_nights']+df['stays_in_week_nights']
df['Revenue']=df['total stay in nights']*df['adr']
```

```python
df.shape
```

### Droping the unwanted features

```python
zero_guests = df[(df['adults'] == 0) & (df['children'] == 0) & (df['babies'] == 0)]
```

- selecting rows where the value in the 'adults' column is 0, the value in the 'children' column is 0, and the value in the 'babies' column is 0.

```python
print(zero_guests.shape)
```

- These 180 rows are considered as unvalid

```python
df = df.drop(zero_guests.index)
```

```python
df.shape
```

- Dropped 180 rows from the dataset

```python
columns_to_drop=['agent','company','children','babies','adults','stays_in_week_nights','stays_in_weekend_nights',
                 'arrival_date_week_number','arrival_date_day_of_month','reservation_status_date']
df = df.drop(columns_to_drop,axis=1)
```

### 1.Find out the most preferred hotel

### 13. Analyzing average daily rate by hotel

```python
avg_daily_rate = df.groupby('hotel')['adr'].mean().reset_index()
avg_daily_rate
```

```python
sns.barplot(x="hotel", y="adr", data=avg_daily_rate, palette="Blues")
plt.xlabel("Hotel")
plt.ylabel("Average daily rate")
plt.title("Average daily rate by Hotel")
plt.show()
```

- The analysis revealed that the City Hotels are more popular among guests with double the number of bookings compared to Resort hotels. However, despite the higher number of bookings, the adr generated by City hotel shows less difference.

### 14.Which month sees the most hotel bookings in each hotel?

```python
colours = ["lightcoral", "indianred"]
sns.countplot(x='arrival_date_month', hue = 'hotel', data= df,palette=colours)
plt.title('Months of Arrival')
plt.xticks(rotation=90)
plt.show()
```

- According to the above analysis, August is the most arrival month for both the City Hotel and the Resort Hotel. This means that more guests checked into these hotels during the month of August than any other month of the year. This observation suggests that August is a peak travel season for both hotels.Hotel operators may need to increase staff and resources to handle the influx of guests, while also offering special promotions and packages to attract more visitors during this peak season.

### 16.Lead time comparison between city and resort hotels
